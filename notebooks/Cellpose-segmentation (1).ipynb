```python
import numpy as np
from skimage.io import imread
import imageio.v2 as iio
import matplotlib.pyplot as plt
import skimage
from skimage.segmentation import mark_boundaries
from skimage.exposure import rescale_intensity
import matplotlib.patches as patches
from skimage.measure import label, regionprops_table
from cellpose import io, models
from cellpose.io import imread

print("imports ok")

#prepping everything and load the required packages
```

```python
# Define the main path

path = '/Volumes/Gabriel_SSD/HAC15_TATDN2/250604_Hac15_TATDN2/segmentation_input/avg/'
```

```python
#Setup Cellpose and read images

io.logger_setup()

# Define the model
model = models.CellposeModel(gpu=True)

# Read image files
img0 = iio.imread(path+'AVG_seg_250604_Hac15__MMStack_Default.tif')
img1 = iio.imread(path+'AVG_seg_250604_Hac15__MMStack_Default.tif')
img2 = iio.imread(path+'AVG_seg_250604_Hac15__MMStack_Default.tif')
```

```python
# Segmentation

masks0, flows0, styles0 = model.eval(img0)
masks1, flows1, styles1 = model.eval(img1)
masks2, flows2, styles2 = model.eval(img2)
```
