```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('../human_adrenals.mplstyle')
```

```python
transient_data = pd.read_csv("Fura-2_transients.csv")
transient_data.decay = transient_data.decay * -1
```

```python
fig, ax = plt.subplots(1,4, figsize=(10, 5))

sns.violinplot(data=transient_data, y="peak_interval", ax=ax[0], fill=False, inner="quarts")
sns.swarmplot(data=transient_data, y="peak_interval", ax=ax[0], color="black", dodge=True, size=2)
ax[0].set_yticks(range(0,3501, 500), range(0,351,50))
ax[0].set_ylabel("Peak-to-Peak Interval (s)")
ax[0].set_ylim(0,3500)

sns.violinplot(data=transient_data, y="fwhm", ax=ax[1], fill=False, inner="quarts")
sns.swarmplot(data=transient_data, y="fwhm", ax=ax[1], color="black", dodge=True, size=2)
ax[1].set_yticks(range(0,501, 100), range(0,51,10))
ax[1].set_ylabel("Full Width Half Maximum (FWHM) (s)")
ax[1].set_ylim(0,500)

sns.violinplot(data=transient_data, y="rise", ax=ax[2], fill=False, inner="quarts")
sns.swarmplot(data=transient_data, y="rise", ax=ax[2], color="black", dodge=True, size=2)
ax[2].set_yticks(range(0,501, 100), range(0,51,10))
ax[2].set_ylabel("10-90% Rise Time (s)")
ax[2].set_ylim(0,500)

sns.violinplot(data=transient_data, y="decay", ax=ax[3], fill=False, inner="quarts")
sns.swarmplot(data=transient_data, y="decay", ax=ax[3], color="black", dodge=True, size=2)

ax[3].set_yticks(range(0,501, 100), range(0,51,10))
ax[3].set_ylabel("90-10% Decay Time (s)")
ax[3].set_ylim(0,500)

plt.tight_layout()
plt.savefig("Fura-2_Transient_Analysis.svg")
plt.show()
```
